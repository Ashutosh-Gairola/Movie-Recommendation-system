```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
merge_titles_credits_update['TAGS']=merge_titles_credits_update['TAGS'].apply(stem)
```

similarity check against tags bags of words (text_vectorization) [(tfidf) (word to vec)]

```python
# combine all the text
# word having maximum frequency ,we will extract those words
# dont consider stop words eg. (are and of to from ) and many more
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=6000,stop_words='english')
```

```python
vectors = cv.fit_transform(merge_titles_credits_update['TAGS']).toarray()
```

```python
vectors[0]
```

```python
len(cv.get_feature_names_out())
```

calculate cosine distance

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(title):
    index = merge_titles_credits_update[merge_titles_credits_update['title'] == title].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
        print(merge_titles_credits_update.iloc[i[0]].title)
```

```python
recommend('The Avengers') # top 5 movies similar will be shown
```
